### Merge/Join
pd.merge(left=df1, right=df2, on= 'colname', how= 'type of merge)
- we use the common column (on=) in both dfs to join/merge them
    * inner(how=): returns rows that exist in both dfs
    * outer: all the values will be joined but if we don't have match then it passes NaN
    * right: produce rows from both dfs where there is a match and if not passes NaN / it first pass the rows that are similar in both and then only add the rows from right df  
    * left: does exactly the opposite of right

```python
left_frame = pd.DataFrame({'key': range(5),
                           'left_value': ['a', 'b', 'c', 'd', 'e']})
right_frame = pd.DataFrame({'key': range(2, 7),
                           'right_value': ['f', 'g', 'h', 'i', 'j']})
print(left_frame)
print('\n')
print(right_frame)
```

```python
pd.merge(left_frame, right_frame, on='key', how='inner') # only the rows that are identical in both dfs
```

```python
pd.merge(left_frame, right_frame, on='key', how='right')
#keep the rows that are identical in left and right then stack the rest from right df and passes NaN for values
# that have not any identical value in the left here 5 and 6 keys
```

```python
pd.merge(left_frame, right_frame, on='key', how='left') # opposite of right
```

```python
pd.merge(left_frame, right_frame, on='key', how='outer') # stacl all the rows and passes NaN for non-identical 
# values
```
