```python
raw_data = {'team': ['kiddies', 'kiddies', 'kiddies', 'kiddies', 'FireDogs', 'FireDogs', 'FireDogs', 'FireDogs', 'Mambaas', 'Mambaas', 'Mambaas', 'Mambaas'],
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
        'Lastname': ['Miller', 'Jackson', 'Smith', 'Williams', 'West', 'Jack', 'Ryaner', 'Stone', 'Sloan', 'Piper', 'Roesewood', 'Petaway'],
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['team', 'company', 'Lastname', 'preTestScore', 'postTestScore'])
df
```

```python
team =df['preTestScore'].groupby(df['team']) # in this way you can save the result of groupby as column
team
```

```python
team = df.goupby('preTestScore')['team'] # but this way only works when we have attribute
team
```

```python
list(df['preTestScore'].groupby(df['team'])) # looking at the value within the new groupby column in a long format
```

```python
team.unstack() # you can only use .unstack() in a datafram structure
```

```python
df['preTestScore'].groupby(df['team']).describe() # it gives us statistical summary of a new column
# very important to use in a larger datasets to get an idea of measure of central tendency of groupby data
```

```python
team.mean() # getting an idea of each team pretestscore
```

```python
df['preTestScore'].groupby([df['team'], df['company']]).mean() # we can have even more columns in groupby
```
